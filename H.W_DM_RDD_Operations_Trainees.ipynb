### Read sample1.txt file into RDD and displaying the first 4 elements

```python
rdd_text = sc.textFile('sample1.txt')
rdd_filtered = rdd_text.filter(lambda line: line.strip() != '')
rdd_filtered.take(4)
```

### Count the total number of rows in RDD

```python
total_rows_filtered = rdd_filtered.count()
print(total_rows_filtered)
```

### Create a function to convert the data into lower case and splitting it

```python
transformed_rdd = rdd_text.map(lambda line: line.lower().split())
result = transformed_rdd.collect()
non_empty_result = [words for words in result if words]
non_empty_result
```

### Remove the stopwords from the previous text. i.e. Remove it.

```python
stopwords = ['a','all','the','as','is','am','an','and',
             'be','been','from','had','I','I’d','why','with']
# Hint: you may need use flatMap
```

```python
stopwords_set = set(stopwords)

def remove_stopwords(words):
    return [word for word in words if word not in stopwords_set]

rdd_text = sc.textFile('sample1.txt')
rdd_processed = rdd_text.map(lambda line: line.strip().lower().split())
rdd_flattened = rdd_processed.flatMap(lambda words: remove_stopwords(words))
result = rdd_flattened.collect()
result
```

### Find the words starting with ‘c’

```python
rdd_c_words = rdd_flattened.filter(lambda word: word.startswith('c'))
result_c_words = rdd_c_words.distinct().collect()  
result_c_words
```

### Reduce the data by key and sum it (use the data from the following list)

```python
list = [('JK', 22), ('V', 24), ('Jimin',24), ('RM', 25)
        , ('J-Hope', 25), ('Suga', 26), ('Jin', 27)
       , ('J-Hope', 12), ('Suga', 25), ('Jin', 34)
       , ('JK', 32), ('V', 44), ('Jimin',14), ('RM', 35)]
# Hint: use reduceByKey
```

```python
rdd = sc.parallelize(list)
rdd_reduced = rdd.reduceByKey(lambda x, y: x + y)
result = rdd_reduced.collect()
result
```

### Creat some key value pairs RDDs

```python
rdd1 = sc.parallelize([('a',2),('b',3)])
rdd2 = sc.parallelize([('a',9),('b',7),('c',10)])
```

### Perform Join operation on the RDDs (rdd1,rdd2)

```python
joined_rdd = rdd1.join(rdd2)
result = sorted(joined_rdd.collect(), key=lambda x: x[0], reverse=True)
result
```
