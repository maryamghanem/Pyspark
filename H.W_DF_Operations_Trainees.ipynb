#### Create a new dataframe with a column called HV Ratio that is the ratio of the High Price versus volume of stock traded for a day.

```python
from pyspark.sql.functions import col

df_with_hv_ratio = df.withColumn("HV Ratio", col("High") / col("Volume"))
df_with_hv_ratio.select("HV Ratio").show()
```

#### What day had the Peak High in Price?

```python
from pyspark.sql.functions import max
from datetime import datetime

max_high = df.agg(max("High")).collect()[0][0]
peak_high_row = df.filter(df.High == max_high).select("Date").collect()[0]
peak_high_date = peak_high_row["Date"]

if isinstance(peak_high_date, datetime):
    peak_high_datetime = peak_high_date
else:
    peak_high_datetime = datetime.combine(peak_high_date, datetime.min.time())

peak_high_datetime
```

#### What is the mean of the Close column?

```python
from pyspark.sql.functions import mean

mean_close = df.agg(mean("Close")).show()
mean_close
```

#### What is the max and min of the Volume column?

```python
from pyspark.sql.functions import max, min

volume_stats = df.select(max("Volume"), min("Volume"))
volume_stats.show()
```

#### How many days was the Close lower than 60 dollars?

```python
days_below_60 = df.filter(df["Close"] < 60).count()
days_below_60
```

#### What percentage of the time was the High greater than 80 dollars ?
#### In other words, (Number of Days High>80)/(Total Days in the dataset)

```python
days_high_above_80 = df.filter(df["High"] > 80).count()
total_days = df.count()

percentage_high_above_80 = (days_high_above_80 / total_days) * 100
percentage_high_above_80
```

#### What is the Pearson correlation between High and Volume?

```python
from pyspark.sql.functions import corr

correlation = df.select(corr("High", "Volume")).collect()[0][0]
correlation_df = spark.createDataFrame([(correlation,)], ["corr(High, Volume)"])
correlation_df.show()
```

#### What is the max High per year?

```python
from pyspark.sql.functions import year, max

df_with_year = df.withColumn("Year", year(df["Date"]))
max_high_per_year = df_with_year.groupBy("Year").agg(max("High"))
max_high_per_year.show()
```

#### What is the average Close for each Calendar Month?
#### In other words, across all the years, what is the average Close price for Jan,Feb, Mar, etc... Your result will have a value for each of these months. 

```python
from pyspark.sql.functions import month, avg

df_with_month = df.withColumn("Month", month(df["Date"]))
avg_close_per_month = df_with_month.groupBy("Month").agg(avg("Close"))
avg_close_per_month_sorted = avg_close_per_month.orderBy("Month")
avg_close_per_month_sorted.show()
```

### Group by "Name" column and using sum function on "salary" column

```python
from pyspark.sql.functions import col, sum

test3 = test3.withColumn("salary", col("salary").cast("int"))
salary_sum_per_name = test3.groupBy("Name").agg(sum("salary"))
salary_sum_per_name.show()
```

### Group by "Name" column and using avg function on "salary" column

```python
from pyspark.sql.functions import avg

avg_salary_per_name = test3.groupBy("Name").agg(avg("salary"))
avg_salary_per_name.show()
```

### Group by "Departments" column and using sum function on "salary" column

```python
from pyspark.sql.functions import sum

salary_sum_per_department = test3.groupBy("Departments").agg(sum("salary"))
salary_sum_per_department.show()
```

### Group by "Departments" column and using mean function on "salary" column:

```python
from pyspark.sql.functions import avg

avg_salary_per_department = test3.groupBy("Departments").agg(avg("salary"))
avg_salary_per_department.show()
```

Group by "Departments" column and using count function on "Departments" column:

```python
from pyspark.sql.functions import count

count_per_department = test3.groupBy("Departments").agg(count("Departments"))
count_per_department.show()
```

### Apply agg to using sum function get the total of salaries

```python
from pyspark.sql.functions import sum

total_salary = test3.agg(sum("salary"))
total_salary.show()
```
